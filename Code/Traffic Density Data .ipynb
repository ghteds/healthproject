```python
#Set CSV file locations. May need modification if data moved to other folder.
#We use Table C08301: Means of Transportation to Work to caculate an indicator of traffic density
#Please refer to https://censusreporter.org/tables/C08301/
#Please refer to https://www.socialexplorer.com/data/ACS2015_5yr/metadata/?ds=ACS15_5yr&table=B08301
acs2010_csvlocation = '../Datasets/ACS_10_5YR_B08301.csv'
acs2011_csvlocation = '../Datasets/ACS_11_5YR_B08301.csv'
acs2012_csvlocation = '../Datasets/ACS_12_5YR_B08301.csv'
acs2013_csvlocation = '../Datasets/ACS_13_5YR_B08301.csv'
acs2014_csvlocation = '../Datasets/ACS_14_5YR_B08301.csv'
acs2015_csvlocation = '../Datasets/ACS_15_5YR_B08301.csv'
acs2016_csvlocation = '../Datasets/ACS_16_5YR_B08301.csv'

#Read the CSV's into dataframes
acs2010 = pd.read_csv(acs2010_csvlocation, encoding="ISO-8859-1")[['GEO.display-label','HD01_VD01','HD01_VD02']].rename(columns={"HD01_VD01": "Total_2010", "HD01_VD02": "Cars/Truck/Van_2010"})
acs2011 = pd.read_csv(acs2011_csvlocation, encoding="ISO-8859-1")[['GEO.display-label','HD01_VD01','HD01_VD02']].rename(columns={"HD01_VD01": "Total_2011", "HD01_VD02": "Cars/Truck/Van_2011"})
acs2012 = pd.read_csv(acs2012_csvlocation, encoding="ISO-8859-1")[['GEO.display-label','HD01_VD01','HD01_VD02']].rename(columns={"HD01_VD01": "Total_2012", "HD01_VD02": "Cars/Truck/Van_2012"})
acs2013 = pd.read_csv(acs2013_csvlocation, encoding="ISO-8859-1")[['GEO.display-label','HD01_VD01','HD01_VD02']].rename(columns={"HD01_VD01": "Total_2013", "HD01_VD02": "Cars/Truck/Van_2013"})
acs2014 = pd.read_csv(acs2014_csvlocation, encoding="ISO-8859-1")[['GEO.display-label','HD01_VD01','HD01_VD02']].rename(columns={"HD01_VD01": "Total_2014", "HD01_VD02": "Cars/Truck/Van_2014"})
acs2015 = pd.read_csv(acs2015_csvlocation, encoding="ISO-8859-1")[['GEO.display-label','HD01_VD01','HD01_VD02']].rename(columns={"HD01_VD01": "Total_2015", "HD01_VD02": "Cars/Truck/Van_2015"})
acs2016 = pd.read_csv(acs2016_csvlocation, encoding="ISO-8859-1")[['GEO.display-label','HD01_VD01','HD01_VD02']].rename(columns={"HD01_VD01": "Total_2016", "HD01_VD02": "Cars/Truck/Van_2016"})
```

```python
#Merge 
acs2010_2011 = pd.merge(acs2010, acs2011, on='GEO.display-label', how='outer')
acs2010_2012 = pd.merge(acs2010_2011, acs2012, on='GEO.display-label', how='outer')
acs2010_2013 = pd.merge(acs2010_2012, acs2013, on='GEO.display-label', how='outer')
acs2010_2014 = pd.merge(acs2010_2013, acs2014, on='GEO.display-label', how='outer')
acs2010_2015 = pd.merge(acs2010_2014, acs2015, on='GEO.display-label', how='outer')
acs2010_2016 = pd.merge(acs2010_2015, acs2016, on='GEO.display-label', how='outer').set_index('GEO.display-label')
acs2010_2016.head()
```
